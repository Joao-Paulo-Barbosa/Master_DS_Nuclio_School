```python
import pandas as pd

df_titanic = pd.read_csv(
    filepath_or_buffer='https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv'
)

display(df_titanic)
```

```python
#1. Função que exibe características do DataFrame
def df_charact(x):
    return display("As dimensões do df são", x.shape,
              "O número de elementos do df é ", x.size,
              "As linhas se chamam ", x.index,
              "As colunas se chamam ", x.columns,
              "O número de elementos em cada coluna é ", x.count(),
              "As primeiras 10 linhas são", x.head(10),
              "As últimas 10 linhas são", x.tail(10))

df_charact(df_titanic)
```

```python
#2. Exibir passageiros agrupados de acordo com cada classe

df_class = df_titanic.groupby('Pclass').size()
display (df_class)
```

```python
#3. Exibir linhas pares do DataFrame

df_pairs = df_titanic.iloc[0::2]
display (df_pairs)


```

```python
# 4. Nomes das pessoas na primeira classe em ordem alfabética

df_fclass = df_titanic[df_titanic['Pclass'] == 1]
df_fclass.sort_values('Name')


```

```python
#5. Porcentagem de pessoas que sobreviveram e morreram, agrupadas por gênero 
df_survivors = df_titanic.groupby(['Sex','Survived'])['Survived'].count()*100/df_titanic['Survived'].count()
display (df_survivors)
```

```python
#6. Porcentagem de pessoas que sobreviveram em cada classe, agrupadas por local de embarque 
df_survivors_class = df_titanic.groupby(['Embarked', 'Pclass', 'Survived'])['Survived'].count()*100/df_titanic['Survived'].count()
display (df_survivors_class)

```

```python
#7. Remoção de passageiros com idade desconhecida
df_ageknown = df_titanic.dropna(subset=['Age'])
display (df_ageknown)
```

```python
#8. Média de idade das mulheres em cada classe, agrupadas por local de embarque
df_w_ageknown = df_ageknown[df_ageknown['Sex'] == 'female']
display (df_w_ageknown)
df_w_ageknown_avg = df_w_ageknown.groupby(['Embarked', 'Pclass'])['Age'].mean()
display (df_w_ageknown_avg)
```

```python
#9. Adição de coluna booleana para identificar menores de idade
df_ageknown['Minor'] = df_ageknown['Age'] < 18
display (df_ageknown)
```

```python
#10 Porcentagem de menores e adultos que sobreviveram em cada classe, agrupadas por local de embarque
df_ageknown_minors = df_ageknown.groupby(['Embarked', 'Pclass', 'Survived', 'Minor'])['Survived'].count()*100/df_ageknown['Survived'].count()
display (df_ageknown_minors)
```

O código abaixo é o resultado final da interação com chat GPT, um código que retorna as chances do usuário sobreviver ao naufrágio do titanic de acordo com os fatores: porto em que embarcou, gênero, classe e se é menor.

Em seguida apresento as interações com o chatGPT , até chegar neste código.

```python
def calculate_survival_rate(port, gender, class_, minor):
    # Filter the dataset based on the input parameters
    filtered_df = df_ageknown[(df_ageknown['Embarked'] == port) & (df_ageknown['Sex'] == gender) & (df_ageknown['Pclass'] == class_) & (df_ageknown['Minor'] == minor)]
    # Calculate the total number of passengers matching the input parameters
    total_passengers = len(filtered_df)
    # Calculate the number of survivors matching the input parameters
    total_survivors = len(filtered_df[filtered_df['Survived'] == 1])
    # Calculate the survival rate
    survival_rate = total_survivors / total_passengers * 100
    return survival_rate

# Get the user input
port = input("Enter the port of embarkation (C, Q, or S): ").upper()
gender = input("Enter the gender (male or female): ").lower()
class_ = int(input("Enter the class (1, 2, or 3): "))
minor = input("Enter if the passenger is a minor (True or False): ").lower() == "true"

# Calculate the survival rate based on the input parameters
survival_rate = calculate_survival_rate(port, gender, class_, minor)

# Print the result
print(f"The chance of survival for a {gender} passenger, in class {class_}, embarked at port {port}, {'who is a minor' if minor else 'who is not a minor'}, is {survival_rate:.2f}%.")
```
